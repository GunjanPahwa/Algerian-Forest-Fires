Algerian Forest Fires Dataset

Dataset Info:

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances of each region.

The period from June 2012 to September 2012. The dataset includes 11 attributes and 1 output attribute(class). The 244 instances have been classifies into fire (138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date: (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observation
2. Temp: temperature noon (temparature max) in Celsius degrees: 22 to 42
3. RH: Relative Humidity in %: 21 to 90
4. Ws: Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code(DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index(ISI) index from the FWI system: 0 to 18.5
10. Buildup index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely fire and not fire

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the region

1: "Bejaia Region Dataset"

2: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
#this means ki 122 row tk region ka column banao jisme value 0 daaldo and uske baad uss column mai value 1 daaldo
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
#we need to drop the 122 row as it has the same values as the head
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns #the column classes has extra space lets remove that
```

```python
#fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

##change the required columns as integer data type

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
#what festures have o as an object class, basicaly kaun kaunse features objects hai
```

```python
for i in objects:
    if i!='Classes': #except classes which is the output feature sab kuch float mai convert hojayega
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
##Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
#this means that index wala column would not get saved
```
